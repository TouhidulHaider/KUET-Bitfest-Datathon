```python
df = pd.read_csv('/kaggle/input/bitfest-datathon-2025/train.csv', encoding = "ISO-8859-1")
test = pd.read_csv('/kaggle/input/bitfest-datathon-2025/test.csv')
df.head()
```

```python
train = df.replace([None, 'N/A'], "")
```

```python
train.head()
```

```python
train = train.fillna("")
```

```python
train.head()
```

```python
target = train.iloc[:,34]
train = train.iloc[:,0:34]
train.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

# Renaming column names for ease of access
list1= [i for i in range(train.shape[1])]
new_Index=[str(i) for i in list1]
train.columns= new_Index
train.head(5)
```

```python
# Convertng headlines to lower case
for index in new_Index:
    train[index]=train[index].str.lower()
train.head(1)
```

```python
' '.join(str(x) for x in train.iloc[1,0:34])
```

```python
headlines = []
for row in range(0,len(train.index)):
    headlines.append(' '.join(str(x) for x in train.iloc[row,0:34]))
```

```python
headlines[0]
```

```python
# headlines
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
## implement BAG OF WORDS
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

```python
target = [round(i) for i in target]
```

```python
# implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset, target)
```

```python
## Predict for the Test Dataset
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
# para_data = pd.DataFrame()
# para_data['paragraph'] = headlines
# para_data['matched_score'] = target
# # para_data.to_csv("para_data.csv")
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
test['matched_score'] = predictions
```

```python
matrix=confusion_matrix(test['matched_score'],predictions)
print(matrix)
score=accuracy_score(test['matched_score'],predictions)
print(score)
report=classification_report(test['matched_score'],predictions)
print(report)
```
